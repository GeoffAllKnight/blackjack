```python
import random
from IPython.display import clear_output

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}

playing = True

class Card:
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
    
    def __str__(self):
        return f"{self.rank} of {self.suit}"

class Deck:
    
    def __init__(self):
        self.deck = []  # start with an empty list
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit,rank))
    
    def __str__(self):
        result = ""
        for card in self.deck:
            result += str(card) +", "
        return result[:-2]

    def shuffle(self):
        random.shuffle(self.deck)
        
    def deal(self):
        return self.deck.pop(0)
    
class Hand:
    def __init__(self):
        self.cards = []  # start with an empty list as we did in the Deck class
        self.value = 0   # start with zero value
        self.aces = 0    # add an attribute to keep track of aces
    
    def __str__(self):
        result = ""
        for card in self.cards:
            result += str(card) +", "
        return result[:-2]
    
    def add_card(self,card):
        self.cards.append(card)
        self.value += values[card.rank]
        if card.rank == "Ace":
            self.aces += 1
        if self.aces != 0:
            self.adjust_for_ace()
        
    def adjust_for_ace(self):
        for i in range(self.aces):
            if self.value > 21:
                self.value -= 10
class Chips:
    
    def __init__(self):
        while True:
            try:
                self.total = int(input("How many chips would you like to start with? "))
            except:
                print("Invalid Entry! Please enter a number.")
            else:
                break
        self.bet = 0
        
    def win_bet(self):
        self.total += self.bet
    
    def lose_bet(self):
        self.total -= self.bet

def take_bet():
    
    while True:
        try:
            player_bet = int(input("How much would you like to bet? "))
        except:
            print("That is an invalid bet. Please enter a number!")
        else:
            if player_bet > total_chips.total:
                print("You don't have enough chips!")
                continue
            total_chips.bet = player_bet
            break

def hit(deck,hand):
    
    newcard = deck.deal()
    hand.add_card(newcard)
    show_some(player_hand,dealer_hand)
    
def hit_or_stand(deck,hand):
    global playing  # to control an upcoming while loop
    
    while playing:
        result = input("Would you like to Hit or Stand? ")
        if result == "Hit":
            hit(deck,hand)
            if hand.value >= 21:
                playing = False
            continue
        elif result == "Stand":
            playing = False
        else:
            print("That is not a correct input, please try again")
            continue
            
def show_some(player,dealer):
    
    clear_output()
    
    print(f"The dealer has: One card face down and {str(dealer.cards[1])}")
    print(f"You have {player}")
    print(f"Your current card value is {player.value}")
    
def show_all(player,dealer):
    
    clear_output()
    
    print(f"The dealer has {dealer}")
    print(f"The dealer's card value is {dealer.value}")
    print(f"You have {player}")
    print(f"You current card value is {player.value}")
    
def player_busts(chips):
    chips.lose_bet()
    print("Womp womp! You busted!")

def player_wins(chips):
    chips.win_bet()
    print("You won!")

def dealer_busts(chips):
    chips.win_bet()
    print("Dealer busted! You won!")
    
def dealer_wins(chips):
    chips.lose_bet()
    print("Womp womp! You lose!")
    
def push(chips):
    chips.bet = 0
    print("Push! You get your money back.")

def replay():
    global playing
    
    playagain = str(input("Would you like to play again? (Yas or Nah): "))

    while playagain != "Yas" and playagain != "Nah":
        playagain = str(input("Wrong input y'all (Yas or Nah): "))
    if playagain == "Yas":
        playing = True
        return True
    return False

#GAME CODE BELOW
print("Welcome to Blackjack!")
total_chips = Chips()
deck = Deck()

while True:  
    # Create & shuffle the deck, deal two cards to each player
    if len(deck.deck) == 0:
        deck = Deck()
    
    player_hand = Hand()
    dealer_hand = Hand()
    
    deck.shuffle()
    player_hand.add_card(deck.deal())
    player_hand.add_card(deck.deal())
    dealer_hand.add_card(deck.deal())
    dealer_hand.add_card(deck.deal())
        
    # Prompt the Player for their bet
    take_bet()
    
    # Show cards (but keep one dealer card hidden)
    show_some(player_hand,dealer_hand)
    
    while playing:  # recall this variable from our hit_or_stand function
        
        # Prompt for Player to Hit or Stand
        hit_or_stand(deck,player_hand)
        
        # Show cards (but keep one dealer card hidden)
        show_some(player_hand,dealer_hand)
        
        # If player's hand exceeds 21, run player_busts() and break out of loop
        if player_hand.value > 21:
            player_busts(total_chips)
            break

    # If Player hasn't busted, play Dealer's hand until Dealer reaches 17
        while dealer_hand.value < 17:
            dealer_hand.add_card(deck.deal())
            
        # Show all cards
        show_all(player_hand,dealer_hand)
        # Run different winning scenarios
    if dealer_hand.value > 21:
        dealer_busts(total_chips)
    elif player_hand.value <= 21 and player_hand.value > dealer_hand.value:
        player_wins(total_chips)
    elif player_hand.value < dealer_hand.value:
        dealer_wins(total_chips)
    elif player_hand.value == dealer_hand.value:
        push(total_chips)
        
    # Inform Player of their chips total 
    print(f"You currently have {total_chips.total} chips left.")
    
    # Ask to play again
    if not replay():
          break
clear_output()
print(f"Congratulations! You walk away with {total_chips.total} chips!")
```
